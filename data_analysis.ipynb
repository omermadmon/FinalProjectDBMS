Upload dow jones data to elasticsearch for further analysis.

```
# to databricks:

# File location and type
file_location = "/FileStore/tables/DJI.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_dj = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df_dj)
```

<h2>Comparing number of tweets in latin america within the two periods<h2>

We would now like to test the hypothesis that claims the amount of tweets is related to the level of morbidity in the country.<br>
Given what we already know about the situation in latin america, we have decided to performe thiscomparison on latin american countries.

```
from pyspark.sql.types import *

latin_america_list = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela', 'Mexico', 'Guatemala']
latin_america_list = [[x] for x in latin_america_list]
schema = StructType([StructField("country", StringType(), True)])
latin_america_df = sqlContext.createDataFrame(latin_america_list, schema=schema)
df_loc = df.withColumn("country", F.explode(F.array(df.location).getItem("country"))).join(latin_america_df, 'country')

df_loc = df_loc.groupBy("label", "country").count()
```

```
labels_count_df = df.groupBy("label").count().withColumnRenamed("count", "tweets_count")
labels_count_records = labels_count_df.select("tweets_count").collect()
n_new = labels_count_records[0].tweets_count
n_old = labels_count_records[1].tweets_count

@udf
def country_ratio(count, label):
  if label == 'old': return count/n_old
  else: return count/n_new
  
df_loc = df_loc.withColumn('country_ratio', country_ratio('count', 'label'))
display(df_loc.sort(F.desc('label')))
```

Based on the query results above, it seems like the increasing in corovavirus cases in latin america increases the amount of tweets in the country.

<h1>Q2<h1>

<h2>Distribution of source devices for each popularity group (out of 5 most used devices)<h2>

We would like to check if there is a relation between the users popularity and the device (source) of it's tweets.<br>
Therefore we have defined a popularity measure: the **average** of the **normallized** followers count, firends count, and listed count.

```
# create normallized popularity measure:

max_fol = df.agg({"user_followers_count": "max"}).collect()[0][0]
max_frd = df.agg({"user_friends_count": "max"}).collect()[0][0]
max_lstd = df.agg({"user_listed_count": "max"}).collect()[0][0]
min_fol = df.agg({"user_followers_count": "min"}).collect()[0][0]
min_frd = df.agg({"user_friends_count": "min"}).collect()[0][0]
min_lstd = df.agg({"user_listed_count": "min"}).collect()[0][0]
```

```
from pyspark.sql.types import *

def popularity(fol, frd, lstd):
    fol_n = (fol - min_fol) / (max_fol - min_fol)
    frd_n = (frd - min_frd) / (max_frd - min_frd)
    lstd_n = (lstd - min_lstd) / (max_lstd - min_lstd)
    return (fol_n + frd_n + lstd_n)/3

popularity_udf = F.udf(popularity, DoubleType())
  
df_pop = df.withColumn('popularity', popularity_udf(F.col('user_followers_count'), F.col('user_friends_count'), F.col('user_listed_count')))
```

Creating four equal groups of popularity based on the popularity measure quantiles - where 4 is the labels of the most popular users and 1 is the least popular group.

```
# compute quantiles
q1,q2,q3 = df_pop.approxQuantile('popularity', [0.25,0.5,0.75],0.01)

def quan(x):
  if x<q1:
    return 1
  if x<q2:
    return 2
  if x<q3:
    return 3
  return 4

quan_udf = F.udf(quan, IntegerType())

src = df.groupBy('source').count().sort(F.desc("count")).limit(5).drop('count')
df_pop = df_pop.join(src, 'source', 'inner')

df_pop = df_pop.withColumn('quan', quan_udf(F.col('popularity')))
df_pop = df_pop.groupBy('quan', 'label','source').count()

display(df_pop.sort('quan'))
```

Observations:
* The amount of web tweets increases as the popularity level grows. 
* The gap between Android and iPhone tweets is narrowing as popularity increases.

Possible explanations:
* The most popular users these days (or maybe in general) might be official news channels or politicians, who tend to use PCs and laptops rather than mobile.
* iPhone is more expensive, and it is reasonable to assume that popular users are richer.

<h2>Keywords usage in the two periods<h2>

Recall that our two periods are small samples represanting the times at the very beginning of the coronavirus spreading, and the time after the 1st and 2nd waves.
<br>We wanted to check the usage in certain words in tweets in these samples, and try to deduce some insights and conclusions out of the results.

```
import re

@udf
def clean(s):
  # remove punc + lowercase text
  res = re.sub(r'[^\w\s]',' ',s)
  final_res = re.sub(r'[\n\r\t]',' ',res)
  return final_res.lower()

#clean text
df2 = df.withColumn('text', clean(F.col('text')))
words = ['flight', 'vaccine', 'sport' , 'protest' , 'mask', 'quarantine']

#count tuples for each period
labels_count_df = df.where(df.lang == 'en').groupBy("label").count().withColumnRenamed("count", "tweets_count")
labels_count_records = labels_count_df.select("tweets_count").collect()
n_new = labels_count_records[0].tweets_count
n_old = labels_count_records[1].tweets_count

# find tuples which have certain words in their text
df3 = df2.where(F.col('text').like("% test %") | F.col('text').like("% tests %")) \
                .withColumn('word', F.lit('test'))


for word in words:
  df4 = df2.where(F.col('text').like("%"+word+"%")).withColumn('word', F.lit(word))
  df3 = df3.union(df4)

df3 = df3.groupby('label','word').count()
  
@udf
def text_ratio(count, label):
  if label == 'old': return count/n_old
  else: return count/n_new
  
df3 = df3.withColumn('ratio', text_ratio('count', 'label'))
display(df3.sort(F.desc('label')))
```

We can clearly see that some words are much more popular in tweets these days then before:
<br>The word *protest*, for example, is now a lot more mentioned in tweets, because of big protests all over the world:
<br> <a href="https://www.ynet.co.il/article/HyhMwaYew">in Israel</a>
<br> <a href="https://en.wikipedia.org/wiki/George_Floyd_protests">in US</a>
<br>Other words, mostly corona-related such as 'mask' and 'quarantine', are in less usage than before.

And as for coronavirus itself...

```
df5 = df2.where(F.col('text').like("%coronavirus%"))\
                .withColumn('word', F.lit('coronavirus'))
df5 = df5.groupby('label','word').count()
df5 = df5.withColumn('ratio', text_ratio('count', 'label'))
display(df5.sort(F.desc('label')))
```
